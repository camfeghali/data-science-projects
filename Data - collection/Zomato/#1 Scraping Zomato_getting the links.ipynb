```python
import requests, bs4
import time
import openpyxl
import pprint
from openpyxl.utils import get_column_letter, column_index_from_string
wb = openpyxl.load_workbook('Zomato_Beirut.xlsx')
sheet = wb['Sheet1']
```

```python
restaurant_list=[]
page_number=1

while page_number <=403:
    url='https://www.zomato.com/beirut/restaurants?page='+str(page_number)
    headers = {
        'user-agent': 'Mozilla/5.0 (Windows NT 10.0; WOW64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/68.0.3440.59 Safari/537.36 Avast/68.0.746.60'}
    cookies = dict(cookies_are='working')
    r = requests.get(url,headers=headers, cookies=cookies, timeout=10)
    soup = bs4.BeautifulSoup(r.text)
    for restaurant in soup.find_all("div", class_="col-s-12"):
        restaurant=str(restaurant).split("ResCard_Name")[1].split("title")[0].split("href=")[1]
        restaurant_list.append(restaurant)
    time.sleep(3)
    page_number=page_number+1
    print(page_number)
```

```python
i=1
for resto in restaurant_list:
    sheet.cell(i,1).value = resto.replace('"',"")
    i=i+1 
```

```python
wb.save('Zomato_Beirut.xlsx')
```

```python
i=0
added_restaurant_list=[]
for cellObj in sheet['A']:
    link=cellObj.value
    url=link
    headers = {
            'user-agent': 'Mozilla/5.0 (Windows NT 10.0; WOW64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/68.0.3440.59 Safari/537.36 Avast/68.0.746.60'}
    cookies = dict(cookies_are='working')
    r = requests.get(url,headers=headers, cookies=cookies, timeout=45)
    soup = bs4.BeautifulSoup(r.text)
    box_general3=soup.find("div", class_="row ui segment")
    
    if "See" and "more" and "locations" in str(box_general3):
        tempUrl=box_general3.find_all("div", class_="clearfix res-main-address-links")
        tempUrl=str(tempUrl).split("href=")[1].split(" ")[0].replace('"',"")
        headers = {
            'user-agent': 'Mozilla/5.0 (Windows NT 10.0; WOW64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/68.0.3440.59 Safari/537.36 Avast/68.0.746.60'}
        cookies = dict(cookies_are='working')
        r = requests.get(tempUrl,headers=headers, cookies=cookies, timeout=10)
        soup = bs4.BeautifulSoup(r.text)
        for restaurant in soup.find_all("div", class_="col-s-12"):
            restaurant=str(restaurant).split("ResCard_Name")[1].split("title")[0].split("href=")[1].replace('"',"").replace(" ","")
            added_restaurant_list.append(restaurant)
    i=i+1
    time.sleep(5)
    print(i)
```
