```python
from bs4 import BeautifulSoup
import openpyxl
from openpyxl.utils import get_column_letter, column_index_from_string
wb = openpyxl.load_workbook('Zawarib.xlsx')
sheet = wb['Sheet1']
```

```python
soup = BeautifulSoup(open("Zawarib page source.txt").read())
```

```python
soup=str(soup)
```

```python
locations = soup.split("myLocations")[1].split("myStories")[0].split('["')[1:]
```

```python
i=2
for location in locations:
    name = location.split('"')[0]
    sheet.cell(i,1).value = name
    
    lat = location.split(",")[1]
    sheet.cell(i,3).value = lat
    
    lon = location.split(",")[2]
    sheet.cell(i,4).value = lon
    
    location_type = location.split(",")[-2].replace('"',"").replace("]","")
    sheet.cell(i,2).value = location_type
    
    i=i+1
    print(name + " "+lat+ " "+lon+" "+location_type)
```

```python
wb.save('Zawarib.xlsx')
```
