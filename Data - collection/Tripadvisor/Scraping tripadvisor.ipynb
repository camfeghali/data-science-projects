```python
from selenium import webdriver
import requests, bs4
import time
import openpyxl
import pprint
from openpyxl.utils import get_column_letter, column_index_from_string
wb = openpyxl.load_workbook('TripAdvisor_Beirut.xlsx')
sheet = wb['Sheet1']
```

```python
lebanon_locations_links=[]
restaurant_list=[]
```

```python
# browser = webdriver.Chrome()
link='https://www.tripadvisor.com/Restaurants-g294004-Lebanon.html'
# browser.get(link)
location_finder=requests.get(link)
soup_location = bs4.BeautifulSoup(location_finder.text)
```

```python
for location in soup_location.find_all("div", class_="geo_name"):
    baseUrl='https://www.tripadvisor.com'
    link=location.a['href']
    url=baseUrl+link
#     print(url)
    lebanon_locations_links.append(url)
```

```python
page_number=20
while page_number<=80:
#     browser = webdriver.Chrome()
    link2="https://www.tripadvisor.com/Restaurants-g294004-oa"+str(page_number)+"-Lebanon.html#LOCATION_LIST"
#     browser.get(link2)
    location_finder2=requests.get(link2)
    soup_location2 = bs4.BeautifulSoup(location_finder2.text)
    locations=soup_location2.find("ul", class_="geoList")
    for a in locations.select('li a'):
        link = a['href']
        baseUrl='https://www.tripadvisor.com'
        url=baseUrl+link
#         print(url)
        time.sleep(0.05)
        lebanon_locations_links.append(url)
    page_number=page_number+20
```

# NAME, STREET NAME, CUISINE, RATING, AMOUNT OF RATINGS, PRICE RANGE, GEOLOCATION

```python
restaurant_list=[]
link=lebanon_locations_links[0]
restaurant_finder=requests.get(link)
soup_restaurant = bs4.BeautifulSoup(restaurant_finder.text)
for restaurant in soup_restaurant.find_all("div", class_="title"):
    try:
        link1=str(restaurant.a)
        link1=link1.split("href=")[1].replace("onclick","lol").split("lol")[0].replace('"',"")
        baseUrl="https://www.tripadvisor.com"
        url=baseUrl+link1
        restaurant_list.append(url)
#         print(url)
    except:'IndexError'
print("FIRST PAGE DONE")

page_number=30
while page_number <= 840:
    base1= "https://www.tripadvisor.com"
    link2 = link.split("-")[0]+"-"+link.split("-")[1]+"-"+"oa"+str(page_number)+"-"+link.split("-")[2]+"#EATERY_LIST_CONTENTS"
    
    restaurant_finder=requests.get(link2)
    soup_restaurant = bs4.BeautifulSoup(restaurant_finder.text)
    for restaurant in soup_restaurant.find_all("div", class_="title"):
        try:
            link1=str(restaurant.a)
            link1=link1.split("href=")[1].replace("onclick","lol").split("lol")[0].replace('"',"")
            baseUrl="https://www.tripadvisor.com"
            url=baseUrl+link1
            restaurant_list.append(url)
            print(url)
        except:'IndexError'    
    page_number=page_number+30         
    time.sleep(5)
```

# STORING AS SOUP OBJECTS

```python
facts_object=[]
reviews_object=[]
name_objects=[]
ratings_object=[]

i=0
for link in restaurant_list:
        ress = requests.get(link)
        time.sleep(0.2)
        soup_finder = bs4.BeautifulSoup(ress.text)
        
        
        ##NAME##
        restoName=soup_finder.find("h1", class_="heading_title")
#         restoName = bs4.BeautifulSoup(restoName.text)
        name_objects.append(restoName)
        
        ###FACTS###
        restoFacts=soup_finder.find("div", class_="ppr_rup ppr_priv_restaurants_detail_info_content")
#         restoFacts = bs4.BeautifulSoup(restoFacts.text)
        facts_object.append(restoFacts)
        
        
        ##RATINGS##
        restoRatings=soup_finder.find("span", class_="overallRating")
#         restoRatings = bs4.BeautifulSoup(restoRatings.text)
        ratings_object.append(restoRatings)
        
        ###REVIEWS###
        restoReviews=soup_finder.find("div", class_="rating_and_popularity")
#         restoReviews = bs4.BeautifulSoup(restoReviews.text)
        reviews_object.append(restoReviews)
        
        print(i)
        i=i+1
        time.sleep(2)
```

```python
i=2
for link in restaurant_list:
    sheet.cell(i,1).value = link
    i=i+1
```

```python
i=2
for name in name_objects:
    name=bs4.BeautifulSoup(name.text)
    name=str(name).split("<p>")[1].split("<")[0]
    sheet.cell(i,2).value = name
    i=i+1
```

```python
price_list=[]
for fact in facts_object:

    fact=bs4.BeautifulSoup(fact.text)
    price=str(fact)
    if "$" in price:
        price=price.split("PRICE")[1][:8].replace("R","").replace("A","").replace("T","").replace("I","").replace("N","").replace("G","").replace("L","")
        price=price.replace("O","").replace("C","").replace("A","").replace("T","").replace("I","").replace("O","").replace("N","")
        price_list.append(price)
    else:
        price_list.append("")
        
i=2
for price in price_list:
    sheet.cell(i,3).value = price
    i=i+1  
```

```python
cuisine_list=[]
for fact in facts_object:
    fact=bs4.BeautifulSoup(fact.text)
    cuisine=str(fact)
    if "CUISINE" in cuisine:
        cuisine=cuisine.split("CUISINE")[1].split("PRICE")[0].split("RATING")[0].split("LOCATION")[0][2:]
        cuisine_list.append(cuisine)
    else:
        cuisine_list.append("")
i=2      
for cuisine in cuisine_list:
    sheet.cell(i,4).value = cuisine
    i=i+1
```

```python
reviews_list=[]
for review in reviews_object:
    review=bs4.BeautifulSoup(review.text)
    review=str(review)
    if "Reviews" in review:
        review=review.split("Reviews")[0].split("<p>")[1]
        reviews_list.append(review)
    else:
        reviews_list.append("")
i=2      
for review in reviews_list:
    sheet.cell(i,5).value = review
    i=i+1
```

```python
lat_lon_list=[]
i=0
for link in restaurant_list:
        ress = requests.get(link)
        time.sleep(0.2)
        soup_finder = bs4.BeautifulSoup(ress.text)
        
        text=str(soup_finder)
        ##GEOLOCATION##
        latlon=text.split("coords")[1].split("}")[0]
        lat_lon_list.append(latlon)
        i=i+1
        print(i)
```

```python
i=2
for latlng in lat_lon_list:
    lat=latlng[3:].split(",")[0]
    lng=latlng[3:].split(",")[1][:-1]
    print(lat + " -- " +lng)
    sheet.cell(i,6).value = lat
    sheet.cell(i,7).value = lng
    i=i+1
```
